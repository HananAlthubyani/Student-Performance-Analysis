**Q1)How do student performances differ between the first and second semesters?**

```python
mean_G1 = df1['G1'].mean()
mean_G2 =df1['G2'].mean()
print(f"Mean of G1 : {mean_G1:.2f}")
print(f"Mean of G2 : {mean_G2:.2f}")
# رسم المقارنة في مخطط أعمدة (Bar Plot)
plt.figure(figsize=(6, 5))
plt.bar(['G1', 'G2'], [mean_G1, mean_G2], color=['blue', 'orange'])
plt.title('the overall average scores between the two periods')
plt.ylim(0, 20) # لتثبيت نطاق الدرجات من 0 إلى 20
plt.show()
```

```python
m_G1 = df1['G1'].median()
m_G2 = df1['G2'].median()

print(f"Median of G1: {m_G1:.2f}")
print(f"Median of G2: {m_G2:.2f}")
```

**Correlationship analysis**

```python
df2 = df1
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2['school'] = le.fit_transform(df2['school'])
df2['sex'] = le.fit_transform(df2['sex'])
df2['address'] = le.fit_transform(df2['address'])
df2['famsize'] = le.fit_transform(df2['famsize'])
df2['guardian'] = le.fit_transform(df2['guardian'])
```

```python
df2.head()
```

```python
corr_matrix = df2.corr()
plt.figure(figsize=(14, 10))

sns.heatmap(
    corr_matrix, 
    annot=True,          
    fmt=".2f",           
    cmap="coolwarm",     
    linewidths=0.5,     
    annot_kws={"size": 8} 
)
plt.xticks(rotation=45, ha='right') 
plt.yticks(rotation=0)

plt.tight_layout() 
plt.show()
```

 Weak positive correlations were observed between G3 and parental education: Medu (r = 0.22) and Fedu (r = 0.15). Family size and address showed very weak positive correlations (r ≈ 0.08 )(r=0.11), while sex showed a weak positive correlation (r = 0.10).
Weak negative correlations were observed for age (r = -0.16), school (r = -0.05), and guardian (r = -0.07).

```python
#select a subset from  feauers of the dataset to be analyzed.
df3= df[['school','sex','age','address','famsize','Medu','Fedu','guardian','G1','G2','G3']]
```

```python
from scipy.stats import ttest_ind

male = df3[df3['sex'] == 'M']['G3']
female = df3[df3['sex'] == 'F']['G3']

t_stat, p_value = ttest_ind(male, female)
print(p_value)
print(t_stat)
```

```python
from scipy.stats import ttest_ind

GP = df3[df3['school'] == 'GP']['G3']
MS= df3[df3['school'] == 'MS']['G3']

t_stat, p_value = ttest_ind(GP,MS)
print(p_value)
print(t_stat)
```

```python
from scipy.stats import ttest_ind

U = df3[df3['address'] == 'U']['G3']
R = df3[df3['address'] == 'R']['G3']

t_stat, p_value = ttest_ind(U,R)
print(p_value)
print(t_stat)
```

```python
from scipy.stats import ttest_ind

GT3 = df3[df3['famsize'] == 'GT3']['G3']
LE3 = df3[df3['famsize'] == 'LE3']['G3']

t_stat, p_value = ttest_ind(GT3,LE3)
print(p_value)
print(t_stat)
```

```python
from scipy.stats import f_oneway

groups = [group['G3'].values for _, group in df3.groupby('guardian')]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
from scipy.stats import f_oneway

groups = [group['G3'].values for _, group in df3.groupby('age')]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
from scipy.stats import f_oneway

groups = [group['G3'].values for _, group in df3.groupby('Fedu')]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
from scipy.stats import f_oneway

groups = [group['G3'].values for _, group in df3.groupby('Medu')]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

Based on the T-test and ANOVA test, we can tell that sex, address, age, Fedu, and Medu give sign fiv affect on the performance of students.
the p-value is less than 0.05
